```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/maria/OneDrive/Escritorio/data.csv",sep=",")
```

```python
data
```

```python
data.columns
data.info
data.describe
```

```python
data.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
regresion=linear_model.LinearRegression()
```

```python
metros=data["metro"].values.reshape((-1,1))
```

```python
modelo=regresion.fit(metros,data["precio"])

print ("Interseccion (b)", modelo.intercept_)
print ("Pendiente (m)", modelo.coef_)
```

```python
entrada= [[5],[10],[15],[20],[25]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
data.plot.scatter(x="metro", y="precio", label='Data')
plt.scatter(entrada, predicciones, color='green')
plt.plot(entrada, predicciones, color='black', label='Linea de Regresion')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
